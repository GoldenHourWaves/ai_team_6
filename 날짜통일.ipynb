```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./st.data/금값 일간 데이터 (250901-251031)_ymd.csv")
df['종가'] = df['종가'].str.replace(',', '').astype(float)
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')
df = df.sort_values('Date')

plt.figure(figsize=(12, 6))

# 1. Y축 범위를 데이터 근처로 바짝 붙여서 변화를 크게 보여줍니다 (ylim 사용)
plt.ylim(df['종가'].min() - 50, df['종가'].max() + 50)

# 2. 그래프 그리기
plt.plot(df['Date'], df['종가'], color='#D4AF37', linewidth=3, marker='o', label='Gold Price')

# 3. 예상하셨던 '치솟는 구간' 강조하기
plt.axvspan(pd.to_datetime('2025-10-10'), pd.to_datetime('2025-10-23'), 
            color='yellow', alpha=0.2, label='Rapid Increase Phase')

# 4. 최고점에 화살표 표시
max_val = df['종가'].max()
max_date = df[df['종가'] == max_val]['Date'].iloc[0]
plt.annotate(f'Peak: {max_val}', xy=(max_date, max_val), xytext=(max_date, max_val+30),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Gold Price Trend: Analysis of October Surge', fontsize=15)
plt.grid(True, axis='y', alpha=0.3)
plt.legend()
plt.show()
```
